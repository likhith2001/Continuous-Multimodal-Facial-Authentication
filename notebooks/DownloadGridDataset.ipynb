```python
import os
import subprocess
import zipfile
import tarfile

# Create required directories
os.makedirs("gridcorpus/raw/audio", exist_ok=True)
os.makedirs("gridcorpus/raw/video", exist_ok=True)
os.makedirs("gridcorpus/audio", exist_ok=True)
os.makedirs("gridcorpus/video", exist_ok=True)

# User input for speaker range and extraction preference
start_speaker = int(input("Enter the starting speaker number: "))
end_speaker = int(input("Enter the ending speaker number: "))
extract_files = input("Do you want to extract files after downloading? (y/n): ").lower()

for i in range(start_speaker, end_speaker + 1):
    print(f"\n\n------------------------- Downloading {i}th speaker -------------------------\n\n")
    
    # Paths for audio and video
    audio_path = f"gridcorpus/raw/audio/s{i}.tar"
    video_path = f"gridcorpus/raw/video/s{i}.zip"

    # Download audio files
    subprocess.run(["curl", f"https://spandh.dcs.shef.ac.uk/gridcorpus/s{i}/audio/s{i}.tar", "-o", audio_path])
    
    # Download video files
    subprocess.run(["curl", f"https://spandh.dcs.shef.ac.uk/gridcorpus/s{i}/video/s{i}.mpg_vcd.zip", "-o", video_path])

    # Extract files if requested
    if extract_files == "y":
        # Extract video files
        if os.path.exists(video_path):
            with zipfile.ZipFile(video_path, 'r') as zip_ref:
                zip_ref.extractall(f"gridcorpus/video/s{i}")
        
        # Extract audio files
        if os.path.exists(audio_path):
            with tarfile.open(audio_path, 'r') as tar_ref:
                tar_ref.extractall(f"gridcorpus/audio/s{i}")

print("Download completed.")
```
